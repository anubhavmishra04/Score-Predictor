```python
import pandas as pd
```

```python
data = pd.read_csv('ODI_ball_by_ball_updated.csv')
```

```python
data = data[data['innings']==1]
data
```

```python
data.drop(columns=['season',
                  'striker',
                  'start_date',
                  'non_striker',
                    'innings'
                  ], inplace=True)
```

```python
data.drop(columns=['bowler'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.fillna(0, inplace=True)
```

```python
data['runs'] = data['runs_off_bat'] + data['extras'] + data['wides'] + data['noballs'] + data['byes'] + data['legbyes'] + data['penalty']
```

```python
data
```

```python
data.drop(columns=['extras',
                  'wides',
                  'runs_off_bat',
                  'noballs',
                  'byes',
                  'legbyes',
                  'penalty'
                  ], inplace=True)
```

```python
data
```

```python
data['batting_team'].unique()
```

```python
teams = [
    'New Zealand',
    'India',
    'Sri Lanka', 
    'Australia', 
    'England',
    'West Indies', 
    'South Africa',
    'Pakistan',
    'Bangladesh',
    'Afghanistan'
]
```

```python
data = data[data['batting_team'].isin(teams)]
data = data[data['bowling_team'].isin(teams)]
```

```python
data
```

```python
eligible_venue = data['venue'].value_counts()[data['venue'].value_counts() > 3000].index.tolist()
```

```python
data = data[data['venue'].isin(eligible_venue)]
```

```python
data
```

```python
data['cur_score'] = data.groupby('match_id').cumsum()['runs']
```

```python
data
```

```python
data.drop(columns=['player_dismissed',
                  'other_player_dismissed'
                  ], inplace=True)
```

```python
data
```

```python
data['over'] = data['ball'].apply(lambda x:str(x).split(".")[0])
data['ball_no'] = data['ball'].apply(lambda x:str(x).split(".")[1])
data
```

```python
data['balls_bowled'] = (data['over'].astype('int')*6) + data['ball_no'].astype('int')
data
```

```python
data['balls_left'] = 300 - data['balls_bowled']
data['balls_left'] = data['balls_left'].apply(lambda x:0 if x<0 else x)
data
```

```python
data['wicket_type'] = data['wicket_type'].apply(lambda x:0 if x==0 else 1)
data['other_wicket_type'] = data['other_wicket_type'].apply(lambda x:0 if x==0 else 1)
data['wicket'] = data['other_wicket_type'] + data['wicket_type']
data
```

```python
data['wickets_left'] = 10 - data.groupby('match_id').cumsum()['wicket']
data
```

```python
data.drop(columns=['wicket_type',
                  'other_wicket_type',
                  'wicket'
                  ], inplace=True)
```

```python
data
```

```python
data['crr'] = (data['cur_score']*6)/data['balls_bowled']
data
```

```python
groups = data.groupby('match_id')

match_ids = data['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
last_five
```

```python
data['last_five'] = last_five
data
```

```python
final_data = data.groupby('match_id').sum()['runs'].reset_index().merge(data, on='match_id')
```

```python
final_df = final_data[['batting_team','bowling_team','venue','cur_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','venue'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
import pickle
pickle.dump(pipe,open('Score_Predictor.pkl','wb'))
```
